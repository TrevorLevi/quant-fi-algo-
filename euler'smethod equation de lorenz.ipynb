equation de lorenz :

x' = 10(y-x)
y' = 28x-y-xz
z' = xy -8/3z
x0,y0,z0


    

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
```

```python
def lorenz(x0,y0,z0,T,p):
    Lx ,Ly ,Lz = [x0],[y0],[z0]
    x,y,z = x0 , y0 ,z0
    for i in range(int(T/p)):
        x = x +p*10*(y-x)
        y = y +p*(28*x-y-x*z)
        z = z+p*(x*y-(8/3)*z)
        Lx.append(x)
        Ly.append(y)
        Lz.append(z)
    dessin = Axes3D(plt.figure(figsize=(10,10)))  
    dessin.plot(Lx,Ly,Lz)
    plt.show()
    
```

```python
lorenz(1,1,1,100,0.001)
```

```python
def Chaos_Pas(x0,y0,z0,T,pas):
    x,y,z = x0,y0,z0
    X,Y,Z = x0,y0,z0
    Ex,Ey,Ez = [0],[0],[0]
    for i in range(1, int(T*10/pas)+1):
        x,y,z = x +pas/10*10*(y-x) ,y +pas/10*(28*x-y-x*z) ,z+pas/10*(x*y-(8/3)*z)
        if i % 10 == 0 :
            X = X +pas*10*(Y-X)
            Y= Y +pas*(28*X-Y-X*Z)
            Z = Z+pas*(X*Y-(8/3)*Z)
            Ex.append(X-x)
            Ey.append(Y-y)
            Ez.append(Z-z)
    dessin = Axes3D(plt.figure(figsize=(10,10)))  
    dessin.plot(Ex,Ey,Ez)
    plt.show()        
        
```

```python
Chaos_Pas(1,1,1,100,0.001)
```
