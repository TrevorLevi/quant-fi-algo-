### buble sort 

-Parcourez la liste de gauche à droite.

-Comparez chaque paire d'éléments adjacents.

-Si un élément est plus grand que le suivant, échangez-les.

-Répétez ce processus pour chaque élément de la liste.

````python
def bubblesort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

````

````python
bubblesort([1,3,7,2,77,34])
````

````python
import random

random_numbers = [random.randint(1, 100000) for _ in range(10000)]
````

````python
bubblesort(random_numbers)
````

### Merge sort 

```python
Divisez la liste en deux moitiés égales (ou presque égales).
Triez chaque moitié récursivement en utilisant le tri par fusion.
Fusionnez les deux moitiés triées pour produire la liste triée finale.
```

````python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        # Fusion des deux moitiés triées
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Vérifiez s'il reste des éléments dans la moitié gauche
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # Vérifiez s'il reste des éléments dans la moitié droite
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr


````

````python
merge_sort([1,3,7,2,77,34])
````

````python
merge_sort(random_numbers)
````

### quick sort 

````python
def quick_sort(arr):
    if len(arr)<=  1:
        return arr
    else :
        pivot = arr[len(arr)//2]
        left = [x for x in arr if x < pivot ]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)
    
    
````

````python
quick_sort([1,3,7,2,77,34])
````

````python
quick_sort(random_numbers)
````

### linear search

````python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
````

````python
# Test data
numbers = [10, 20, 30, 40, 50]
target = 30

# Test the function
index = linear_search(numbers, target)

if index != -1:
    print(f"Target found at index: {index}")
else:
    print("Target not found in list")
````

### Binary Search

````python
def binary_search(arr , target):
    debut = 0 
    fin = len(arr)-1

    while debut <= fin :
        mid = (debut + fin) // 2
        if arr[mid]==target:
            return mid 
        elif arr[mid]<target :
            debut = mid+1
        else :
            fin = mid -1
    return -1
````

### Shortest Path Algorithms







````python
import heapq

def algo_dijkstra(start , graph):
    distances = { noeud : float(10**6)  for noeud in graph}
    print(distances)
    distances[start]=0
    queue= [(0 , start)]

    while queue :
        actuel_distance , actuel_noeud  = heapq.heappop(queue)
        print((actuel_distance , actuel_noeud))

        if actuel_distance > distances[actuel_noeud]:
            continue
        
        for voisin , distance_voisin in graph[actuel_noeud].items():
            distance = actuel_distance + distance_voisin
            print(distance)

            if distance < distances[voisin]:
                distances[voisin ] = distance 
                heapq.heappush(queue , (distance , voisin ))

    return distances
````

### BFS 

````python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph[node] - visited)

    return result

# Exemple d'utilisation

graph = {
    'A': {'B', 'C'},
    'B': {'A', 'D', 'E'},
    'C': {'A', 'F'},
    'D': {'B'},
    'E': {'B', 'F'},
    'F': {'C', 'E'}
}

start_node = 'B'
bfs_result = bfs(graph, start_node)
print(f"BFS traversal from {start_node}: {bfs_result}")
````

### DFS

````python
def dfs_iterative(graph, start):
    visited = set()
    stack = [start]
    result = []

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            result.append(node)
            stack.extend(graph[node] - visited)
    
    return result

# Exemple d'utilisation
dfs_result_iterative = dfs_iterative(graph, start_node)
print(f"DFS traversal (iterative) from {start_node}: {dfs_result_iterative}")
````
